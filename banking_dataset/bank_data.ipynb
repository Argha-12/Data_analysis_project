```python
import pandas as pd

df = pd.read_csv("Comprehensive_Banking_Database.csv")
df.head(5)
# df.isna()
```

```python
# What are the most common types of transactions (e.g., withdrawals, deposits, transfers)
# Count occurrences of each transaction type


transaction_counts = df['Transaction Type'].value_counts()

# Convert the result to a DataFrame for better display
transaction_counts_df = transaction_counts.reset_index()
transaction_counts_df.columns = ['transaction_type', 'count']
transaction_counts_df = transaction_counts_df.sort_values(by='count', ascending=False)

print(transaction_counts_df)
```
